```python
from tkinter import *
import sys

def hello(event):
    print("Single Click, Button-l")
    
def quit(event):
    print("Double Click, so let's stop")
    sys.exit()
    
widget = Button(None, text = 'Mouse Clicks')
widget.pack()

widget.bind('<Button-1>', hello)
widget.bind('<Double-1>', quit)

mainloop()
```

```python
# 연습문제 1
# 버튼 2개 생성 후
# 첫 번째 버튼 한 번 누르면 강아지 사진 보여주기
# 두 번째 버튼 한 번 누르면 고양이 사진 보여주기
```

```python
from tkinter import *
from tkinter import Tk, Canvas
from PIL import ImageTk, Image

root = Tk()
canvas = Canvas(root, width = 400, height = 300)
canvas.pack()

def clicked1():
    im = Image.open('dog.gif')
    canvas.image = ImageTk.PhotoImage(im)
    canvas.create_image(0, 0, image = canvas.image, anchor = 'nw')
    
def clicked2():
    im = Image.open('cat.gif')
    canvas.image = ImageTk.PhotoImage(im)
    canvas.create_image(0, 0, image = canvas.image, anchor = 'nw')
    
but1 = Button(root, text = 'Dog', command = clicked1)
but1.pack()
but2 = Button(root, text = 'Cat', command = clicked2)
but2.pack()
root.mainloop()
```

```python
# 연습문제 2
# 연습문제 1 코드를 수정하여 버튼을 하나만 생성
# 버튼을 한 번 클릭하면 강아지 사진 보여주기
# 버튼을 두 번 클릭하면 고양이 사진 보여주기
```

```python
from tkinter import *
from tkinter import Tk, Canvas
from PIL import ImageTk, Image

root = Tk()
canvas = Canvas(root, width = 400, height = 300)
canvas.pack()

def oneclick(event):
    im = Image.open('dog.gif')
    canvas.image = ImageTk.PhotoImage(im)
    canvas.create_image(0, 0, image = canvas.image, anchor = 'nw')
    
def doubleclick(event):
    im = Image.open('cat.gif')
    canvas.image = ImageTk.PhotoImage(im)
    canvas.create_image(0, 0, image = canvas.image, anchor = 'nw')
    
widget = Button(None, text = 'Mouse Clicks')
widget.pack()

widget.bind('<Button-1>', oneclick)
widget.bind('<Double-1>', doubleclick)

root.mainloop()
```

```python
# color chooser

from tkinter import *
from tkinter.colorchooser import *

def callback():
    result = askcolor(color = "#6A9662", title = "Color Chooser")
    print(result)
    
root = Tk()

Button(root, text = 'Choose Color', fg = "darkgreen", command = callback).pack(side = LEFT, padx = 10)
Button(text = 'Quit', command = root.quit, fg = "red").pack(side = LEFT, padx = 10)

root.mainloop()
```

```python
# Creating menus

from tkinter import *

def NewFile():
    print("New File!")
def OpenFile():
    print("Open File!")
def About():
    print("This is a simple example of a menu")
    
root = Tk()

menu = Menu(root)
root.config(menu = menu)
filemenu = Menu(menu)
menu.add_cascade(label = "File", menu = filemenu)
filemenu.add_command(label = "New", command = NewFile)
filemenu.add_command(label = "Open", command = OpenFile)
filemenu.add_separator()
filemenu.add_command(label = "Exit", command = root.quit)

helpmenu = Menu(menu)
menu.add_cascade(label = "Help", menu = helpmenu)
helpmenu.add_command(label = "About...", command = About)

mainloop()
```

```python
# 연습문제 3
# 그림판 만들어서 색상 선택 가능하게 하기
# 파일 열기, 저장 메뉴도 같이 만들기
```

```python
from tkinter import *
from tkinter.colorchooser import *

def NewFile():
    print("New File!")
def OpenFile():
    print("Open File!")
def About():
    print("This is a simple example of a menu")
    
root = Tk()
menu = Menu(root)
root.config(menu = menu)

filemenu = Menu(menu)
menu.add_cascade(label = "File", menu = filemenu)
filemenu.add_command(label = "New", command = NewFile)
filemenu.add_command(label = "Open", command = OpenFile)
filemenu.add_separator()
filemenu.add_command(label = "Exit", command = root.quit)

helpmenu = Menu(menu)
menu.add_cascade(label = "Help", menu = helpmenu)
helpmenu.add_command(label = "About...", command = About)

canvas = Canvas(root, width = 500, height = 500)
color = "red"
result = "#476042"

def callback():
    global result
    result = askcolor(title = "Color Chooser")
    result = result[1]
    
button = Button(root, text = 'Choose Color', fg = "darkgree", command = callback)
button.pack(side = LEFT, padx = 10)

lastx, lasty = 0, 0

def xy(event):
    global lastx, lasty
    lastx, lasty = event.x, event.y
    
def addLine(event):
    global lastx, lasty
    canvas.create_line((lastx, lasty, event.x, event.y), fill = result)
    lastx, lasty = event.x, event.y

root.columnconfigure(0, weight = 1)
root.rowconfigure(0, weight = 1)

canvas.pack()
canvas.bind("<Button-1>", xy)
canvas.bind("<B1-Motion>", addLine)

root.mainloop()
```
