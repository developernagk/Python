```python
# for_range
# for 반복문과 범위를 함께 조합해서 사용합니다
for i in range(5):
    print(str(i) + "= 반복 변수")
print()

for i in range(5, 10):
    print(str(i) + "= 반복 변수")
print()

for i in range(0, 10, 3):
    print(str(i) + "= 반복 변수")
print()
```

```python
# list_range01
# 리스트를 선언합니다
array = [273, 32, 103, 57, 52]

# 리스트에 반복문을 적용합니다
for i in range(len(array)):
    # 출력합니다
    print("{}번째 반복 : {}".format(i, array[i]))
```

```python
# reversed_for01
# 역반복문
for i in range(4, 0 - 1, -1):
    # 출력합니다
    print("현재 반복 변수 : {}".format(i))
```

```python
# reversed_for02
# 역반복문
for i in reversed(range(5)):
    # 출력합니다
    print("현재 반복 변수 : {}".format(i))
```

```python
# for_pyramid01
output = ""

for i in range(1, 10):
    for j in range(0, i):
        output += "*"
    output += "\n"
    
print(output)
```

```python
output = ""

for i in range(1, 10):
    output += ("*" * i)
    output += "\n"
    
print(output)
```

```python
# for_pyramid02
output = ""

for i in range(1, 15):
    for j in range(14, i, -1):
        output += ' '
    for k in range(0, 2 * i - 1):
        output += '*'
    output += '\n'
    
print(output)
```

```python
# infinite_loop(강제로 멈출 때까지 멈추지 않음)
# while 반복문을 사용합니다
while True:
    # "."를 출력합니다
    # 기본적으로 end가 "\n"이라 줄바꿈이 일어나는데
    # 빈 문자열 ""로 바꿔서 줄바꿈이 일어나지 않게 합니다
    print(".", end="")
```

```python
# while_as_for
# 반복 변수를 기반으로 반복하기
i = 0
while i < 10:
    print("{}번째 반복입니다.".format(i))
    i += 1
```

```python
# while_with_condition
# 변수를 선언합니다
list_test = [1, 2, 1, 2]
value = 2

# list_test 내부에 value가 있다면 반복
while value in list_test:
    list_test.remove(value)
    
# 출력합니다
print(list_test)
```

```python
# while_with_time(시간을 조건으로 걸 때는 while 반복문)
# 시간과 관련된 기능을 가져옵니다
import time

# 변수를 선언합니다
number = 0

# 5초 동안 반복합니다
target_tick = time.time() + 5
while time.time() < target_tick:
    number += 1
    
# 출력합니다
print("5초 동안 {}번 반복했습니다.".format(number))
```

```python
# break
# 변수를 선언합니다
i = 0

# 무한 반복합니다
while True:
    # 몇 번째 반복인지 출력합니다
    print("{}번째 반복문입니다.".format(i))
    i = i + 1
    # 반복을 종료합니다
    input_text = input(">반복을 종료하시겠습니까?(y/n) : ")
    if input_text in ["y", "Y"]:
        print("반복을 종료합니다.")
        break
```

```python
# break01
# 변수를 선언합니다
numbers = [5, 15, 6, 20, 7, 25]

# 반복을 돌립니다
for number in numbers:
    # number가 10보다 작으면 다음 반복으로 넘어갑니다
    if number < 10:
        continue
    # 출력합니다
    print(number)
```
